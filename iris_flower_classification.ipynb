```python
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()

print("Raw Iris Data")
print(iris.feature_names)
print(iris.data)
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots()

unique_targets = set(iris.target)

for target in unique_targets:
    # Get indices of rows with the current target
    indices = [i for i, x in enumerate(iris.target) if x == target]
    
    # Plot the data for the current target
    ax.scatter(iris.data[indices, 0], iris.data[indices, 1], label=iris.target_names[target])

ax.set_xlabel('Sepal Length')
ax.set_ylabel('Sepal Width')
ax.set_title('Iris Data')

ax.legend()

plt.show()



```
