Laden sie zunächst den IMDB-Top-250-Datensatz mit der Pandas-Bibliothek, den sie als CSV-Datei auf Backstage finden.

```python
import pandas as pd

file_path: str = "imdb_top_250.csv"
data_frame: pd.DataFrame = pd.read_csv(file_path)
print(data_frame)
```

Untersuchen sie den resultierenden `DataFrame` mittels `.shape`, `.info()` und `.describe()`, welche Informationen sehen sie?

```python
print(f"Shape: {data_frame.shape}")
print(f"Info:")
print(data_frame.info())
print(f"describe:")
print(data_frame.describe())
```

Lassen sie sich die ersten und die letzten zehn Einträge des Datensatzes anzeigen.

```python
print("Ersten Zehn:")
data_frame.head(10)
```

```python
print("Letzten Zehn:")
data_frame.tail(10)
```

Lassen sie sich die ersten fünf Filme anzeigen und beschränken sie sich dabei auf die Spalten `Title`, `Duration` und `Rating count`.

```python
data_frame[['Title', 'Duration', 'Rating count']].head(5)
```

Lassen sie sich die ersten fünf Filme mit dem Genre `Action` anzeigen, beschränken sie sich dabei auf die Spalten `Title` und `Genre`.

```python
data_frame[['Title', 'Genre']].where(data_frame.Genre.str.contains('Action')).dropna()
```

Sortieren sie die Filme nach ihrem Merkmal `IMDB rating` und lassen sie sich die Top-10-Filme mit Namen und Bewertung anzeigen.

```python
data_frame.sort_values(by='IMDB rating', ascending=False).head(10)[['Title', 'IMDB rating']]
```

Gruppieren sie die Filme nach ihrem Regisseur und lassen sie sich die Top-5-Regisseure bezüglich der Anzahl an Filmen anzeigen.

```python
director_counts = data_frame.groupby('Director').size()
director_counts.sort_values(ascending=False).head(5)
```

Visualisieren sie einen Teil der Daten mit `pandas.plotting.scatter_matrix` und `groupby.DataFrameGroupBy.hist`.

```python
pd.plotting.scatter_matrix(data_frame)
data_frame.groupby('Genre').hist(column='IMDB rating', bins=10, alpha=0.5, figsize=(10, 10))


```
