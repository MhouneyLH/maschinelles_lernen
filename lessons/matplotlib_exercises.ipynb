Gegeben sei folgende Liste von Datenpunkten `dpts = [(3,3),(4,5),(4.5,6),(9,7)]`, erzeugen sie ein Diagramm mit der Funktion `plt.plot(xcoors, ycoors)`.

```python
import matplotlib.pyplot as plt

dpts: list = [(3,3),(4,5),(4.5,6),(9,7)]

xcoors: list = [x for x, y in dpts]
ycoors: list = [y for x, y in dpts]

plt.plot(xcoors, ycoors, '-o')
```

Gegeben sind die zwei Reihungen `c = [(1,2),(3,1),(0,1),(2,2)]` \u0000und `\u0000d = [(12,9),(8,10),(11,11),(14,13)]`, \n   stellen sie diese mit Hilfe eines Scatterplots graphisch dar unter Verwendung der Funktion `plt.scatter(xcoors, ycoors)`, wobei `c` und `d` unterschiedliche Farben haben sollen. Die x-Koordinaten sollen die Zeitdauer eines Einkaufs sein und die y-Koordinaten die beim Einkauf ausgegebene Geldmenge, erzeugen sie auch entsprechende Achsenbeschriftungen.

```python
c: list = [(1,2),(3,1),(0,1),(2,2)]
d: list = [(12,9),(8,10),(11,11),(14,13)]

def getXCoords(the_list: list) -> list:
    return [x for x, y in the_list]

def getYCoords(the_list: list) -> list:
    return [y for x, y in the_list]

# das hier geht auch mit subplots
# Link: https://matplotlib.org/stable/gallery/subplots_axes_and_figures/subplots_demo.html
fig = plt.figure(1)
plt.plot(getXCoords(c), getYCoords(c), '-o', color='red')
plt.plot(getXCoords(d), getYCoords(d), '-o', color='blue')
plt.xlabel('Zeitdauer eines Einkaufs')
plt.ylabel('beim Einkauf ausgegebene Geldmenge')

fig2 = plt.figure(2)
plt.plot(getXCoords(c), getYCoords(c), '-o', color='red')

plt.show()
```

```python
import numpy as np


# Berechnen sie für die zwei Reihungen `c\u0000R\u0000e\u0000i\u0000d` auch die paarweisen Abstände und speichern sie diese in einer Matrix, visualisieren sie die Matrix als Heatmap unter Verwendung der Funktion `plt.imshow(matrix, cmap='coolwarm').


matrix: list = np.zeros((len(c), len(c)))
for i in range(len(c)):
    for j in range(len(c)):
        # ist das gleiche, wie die euklidische Distanz
        # distance: float = distance: float = np.sqrt((c[i][0] - d[j][0])**2 + (c[i][1] - d[j][1])**2)
        distance: float = np.linalg.norm(np.array(c[i]) - np.array(d[j]))
        matrix[i][j] = distance

print(c)
print(d)
print(matrix)

fig3 = plt.figure(3)
plt.imshow(matrix, cmap='coolwarm')
plt.colorbar()
plt.show()
```
